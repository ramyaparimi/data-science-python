# Raingall In India

```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
rainfall = pd.read_csv('./district wise rainfall normal.csv', index_col = 'STATE_UT_NAME')
rainfall.head()
```

```python
#using only annual column of the data
annual_rainfall = rainfall[['ANNUAL']].copy()

#applying high rainfall filter
is_high = annual_rainfall['ANNUAL'] > 1000

#grouping by state name and combining the rainfall by state
high_rainfall = annual_rainfall[is_high]



high_rainfall_aggregate = high_rainfall.groupby('STATE_UT_NAME').aggregate(sum)
high_rainfall_aggregate.head()

#plotting a bar graph, setting labels and title for the plot
high_rainfall_plot = high_rainfall_aggregate.plot(kind='bar', figsize=(30, 10))
plt.xlabel('Amount of Rain', fontsize = 15, weight='bold')
plt.ylabel('State Name', fontsize = 15, weight='bold')
plt.title('State Wise Annual Rainfall in India', fontsize = 15, weight='bold')

# plt.savefig('./high_rainfall_states_India.jpeg')


```
